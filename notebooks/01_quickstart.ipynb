# Football Prediction v1.9 Quickstart
Train, predict and inspect the v1.9 rule overlay.


```python
import pandas as pd
from football_prediction_v19.model import train_from_matches, save_model, predict_feature_rows
from football_prediction_v19.features import build_fixture_features
from football_prediction_v19.rules_v19 import assess_prediction
```

```python
matches = pd.read_csv('../data/sample_matches.csv')
model, table, metrics, cols = train_from_matches(matches, test_season=2023)
metrics
```

```python
fixture = build_fixture_features(matches, 'Chelsea', 'Arsenal', '2024-05-01', venue='Stamford Bridge', referee='Anthony Taylor', odds_home=2.4, odds_draw=3.4, odds_away=2.9)
bundle = {'model': model, 'feature_cols': cols, 'metrics': metrics}
pred = predict_feature_rows(bundle, fixture).iloc[0]
assess_prediction(pred, {'H': pred.prob_home, 'D': pred.prob_draw, 'A': pred.prob_away})
```
